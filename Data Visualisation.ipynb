# Data Visualization

```python
plt.figure(figsize = (10,6))
ax=sns.countplot(x = 'employment_type',data = data, palette = 'inferno', legend=False, hue='employment_type')
plt.xlabel('employment_type')
plt.ylabel('Count')
plt.title('No.of employees working with the employment type', fontsize=15,fontweight='bold')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(y ='salary', x = 'experience_level', data = data, palette = 'inferno', legend=False, hue='experience_level')
plt.title('salary based on experience level')
```

```python
plt.figure(figsize = (10,6))
ax=sns.countplot(x = 'salary_currency',data = data, palette = 'inferno', legend=False, hue= 'salary_currency')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

```python
plt.figure(figsize = (10,6))
plt.xticks(rotation= 90)
sns.boxplot(y ='salary', x = 'job_category', data = data, palette = 'rainbow', legend =False, hue='job_category')
plt.title('Salary by Job Category', fontsize=15,fontweight='bold')
```

```python
plt.figure(figsize=(15,7))
chains=data['job_title'].value_counts()[:20]
ax=sns.barplot(x=chains.index,y=chains,palette='Set1', legend = False, hue=chains.index)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 3., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.xticks(rotation=90)
plt.title("No of employees working with the job title ",size=20,pad=20)
plt.xlabel("job title",size=15)
```

```python
plt.figure(figsize=(15,7))
chains=data['company_location'].value_counts()[:20]
ax=sns.barplot(x=chains.index,y=chains,palette='Set1', legend = False, hue=chains.index)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 3., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.xticks(rotation=90)
plt.title("No of employees working in different company locations",size=20,pad=20)
plt.xlabel("Number of locations",size=15)
```

```python
plt.figure(figsize=(15,7))
chains=data['salary'].value_counts()[:20]
ax=sns.barplot(x=chains.index,y=chains,palette='rainbow', legend = False, hue=chains.index)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 3., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title("No of employees having various salaries",size=20,pad=20)
plt.xlabel("salary",size=15)
```

```python
sns.set(font_scale = 0.7)
plt.figure(figsize=(10,7))
labels=['2023', '2022', '2021', '2020']
chains=data['work_year'].value_counts()[:20]
chart=plt.pie(chains, labels=labels, autopct='%1.1f%%', startangle=180, colors=sns.color_palette('rainbow'))
plt.title('Employee working year', fontsize=15,fontweight='bold')
plt.show()
```
